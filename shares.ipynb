```python
class Share:
    def __init__(self, miner_id, difficulty, timestamp):
        self.miner_id = miner_id
        self.difficulty = difficulty
        self.timestamp = timestamp

# Example input data
block_reward = 10.0  # Reward for finding a block
block_difficulty = 1000000  # Block difficulty
N = 2  # Multiplier for PPLNS
total_difficulty_threshold = N * block_difficulty

# List of shares (in reverse chronological order)
shares = [
    Share('miner1', 100000, 1650000000),
    Share('miner2', 150000, 1650000100),
    Share('miner3', 200000, 1650000200),
    Share('miner1', 300000, 1650000300),
    Share('miner2', 250000, 1650000400),
    Share('miner3', 500000, 1650000500),
    Share('miner1', 600000, 1650000600),  # Share that found the block
]

# Accumulate total difficulty of shares
total_difficulty = 0
reward_distribution = {}
for share in shares:
    if total_difficulty >= total_difficulty_threshold:
        break
    total_difficulty += share.difficulty
    if share.miner_id not in reward_distribution:
        reward_distribution[share.miner_id] = 0
    reward_distribution[share.miner_id] += share.difficulty

# Calculate the proportional rewards
total_accumulated_difficulty = sum(reward_distribution.values())
payouts = {miner_id: (difficulty / total_accumulated_difficulty) * block_reward
           for miner_id, difficulty in reward_distribution.items()}

# Output the payouts
for miner_id, payout in payouts.items():
    print(f'Miner {miner_id} earns {payout:.6f} BTC')

# Result example:
# Miner miner1 earns X.XXXXXX BTC
# Miner miner2 earns X.XXXXXX BTC
# Miner miner3 earns X.XXXXXX BTC
```
